Recaman sequence for generative art
https://simpleprogrammer.com/python-generative-art-math


```python
def recaman(n):
    rec = []
    rec.append(0)
    for i in range(1,n):
        x = rec[i-1] - i
        if x < 0 or x in rec:
            x = rec[i-1]+ i
        rec.append(x)
    return rec
```

```python
recaman(20)
```

```python
import matplotlib.pyplot as plt
from random import random
from random import sample
```

```python
c1 = [0 , 0, 0, 1]
x = recaman(100)
y = [ rec/max(x) for rec in x ]
plt.figure(figsize=(10,10))
for rad in y:
    a,b = 2*random(), 2*random()
    cen = tuple(sample(c1,2))
    circle=plt.Circle(cen,rad, color ='r', fill = 0 )
    plt.gcf().gca().add_artist(circle)
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt

fg, ax = plt.subplots(1, 1)
y = recaman(100)
x = [ r/max(y) for r in y ]
```

```python
endpoints = [(a,b) for a,b in zip(x,x[1:])]
centrepoint = [(x+y)/2 for x,y in endpoints ]
heights = [(x-y)/2 for x,y in endpoints]
```

```python
import matplotlib.patches as mpatches
import matplotlib.pyplot as plt
fg, ax = plt.subplots(1, 1)
for i,h in enumerate(heights):
    c = centrepoint[i]
    pac = mpatches.Arc((c,0), h, h, angle=0, theta1=0, theta2=180)
    ax.add_patch(pac)

ax.axis([0, 100, -20, 20])
ax.set_aspect("equal")
fg.canvas.draw()
```
